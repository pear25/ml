```python
import numpy as np
import matplotlib.pyplot as plt


rng = np.random.default_rng(seed=42)

n = 50

x = rng.uniform(low=-5, high=5, size=n)

gaussian_noise = rng.normal(loc=0, scale=1, size=n)

y = 3 * x + 2 + gaussian_noise

print(y)
```

```python
plt.scatter(x, y, alpha=0.6, label="data")
x_base = np.linspace(x.min(), x.max(), n)
y_base = 3 * x_base + 2
plt.plot(x_base, y_base, color="red", label="y = 3x + 2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# MSE loss function 

# print(y_base, y)
y_pred = 3 * x + 2
# print(f"y_base: {y_base}")
# print(f"y: {y_pred}")
# print(y_base - y)
mse_loss = np.mean(((y_pred - y) ** 2))
print(mse_loss)
```

```python
# Gradient descent
import numpy as np

w = 50
b = 50

steps = 15

learning_rate = 0.1


y_pred = w * x + b
mse_loss = np.mean((y_pred - y) ** 2)

def derivative_w(w, b, x, y):
    return 2 * np.mean((w * x + b - y) * x)

def derivative_b(w, b, x, y):
    return 2 * np.mean(w * x + b - y)

losses = []

for i in range(steps):
    mse_loss = np.mean((y_pred - y) ** 2)
    losses.append(mse_loss)
    min_w = derivative_w(w, b, x, y)
    min_b = derivative_b(w, b, x, y)
    w = w - learning_rate * min_w
    b = b - learning_rate * min_b
    
    print(f"new w: {w}")
    print(f"new b: {b}")

    y_pred = w * x + b

print(f"final w: {w}")
print(f"final b: {b}")

y_pred = w * x + b

mse_loss = np.mean((y_pred - y) ** 2)

print(f"final mse loss: {mse_loss}")
```

```python
plt.plot(losses)
plt.xlabel("step")
plt.ylabel("MSE")
# plt.yscale("log")
# print(losses)
```
